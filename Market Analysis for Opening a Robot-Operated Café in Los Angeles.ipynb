##  Análisis de datos

### Proporciones de los distintos tipos de establecimientos

```python
# Calcular proporciones de los distintos tipos de establecimientos
type_proportions = data['object_type'].value_counts(normalize=True)

# Graficar proporciones
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=type_proportions.index, y=type_proportions.values, palette="viridis")
plt.title('Proporciones de los distintos tipos de establecimientos')
plt.xlabel('Tipo de establecimiento')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.show()
```

### Tipo de establecimiento habitualmente una cadena

```python
# Agrupar por tipo de establecimiento y cadena, y contar
chain_by_type = data.groupby(['object_type', 'chain']).size().unstack(fill_value=0)

# Calcular proporción de cadenas por tipo de establecimiento
chain_by_type_proportion = chain_by_type.div(chain_by_type.sum(axis=1), axis=0)

# Mostrar resultados
chain_by_type_proportion
```

### Promedio de número de asientos para cada tipo de restaurante

```python
# Calcular el promedio de número de asientos para cada tipo de restaurante
average_seats_by_type = data.groupby('object_type')['number'].mean()

# Graficar promedio de asientos por tipo de restaurante
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x=average_seats_by_type.index, y=average_seats_by_type.values, palette="viridis")
plt.title('Promedio de número de asientos por tipo de restaurante')
plt.xlabel('Tipo de restaurante')
plt.ylabel('Promedio de número de asientos')
plt.xticks(rotation=45)
plt.show()
```

### Diez mejores calles por número de restaurantes

### Distribución del número de asientos en calles con muchos restaurantes

```python
# Extraer nombres de las calles quitando los números de las direcciones
data['street'] = data['address'].apply(lambda x: ' '.join(x.split()[1:]))

# Seleccionar las 10 calles con más restaurantes
top_10_streets = data['street'].value_counts().head(10).index

# Filtrar datos para estas calles
data_top_10_streets = data[data['street'].isin(top_10_streets)]

# Graficar distribución del número de asientos
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))
sns.boxplot(x='street', y='number', data=data_top_10_streets, palette="viridis")
plt.title('Distribución del número de asientos en las 10 calles con más restaurantes')
plt.xlabel('Calle')
plt.ylabel('Número de asientos')
plt.xticks(rotation=45)
plt.show()
```
